```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv(r'C:\dta\book21.csv',encoding='latin-1')
```

```python
book.head()
```

```python
 book1=book.drop('Unnamed: 0', axis=1)    
```

```python
book1.rename(columns={'User.ID': 'userid'}, inplace=True)
book1.rename(columns={'Book.Title':'Booktitle'},inplace=True)
book1.rename(columns={'Book.Rating':'Bookrating'},inplace=True)
```

```python
#number of unique users in the dataset
len(book1.userid.unique())
```

```python
len(book1.Booktitle.unique())
```

```python
book1.head()
```

```python
user_book = book.pivot_table(index='userid',
                                 columns='Booktitle',
                                 values='Bookrating').reset_index(drop=True)
```

```python
user_book
```

```python
#Impute those NaNs with 0 values
user_book.fillna(0, inplace=True)
```

```python
user_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book1.userid.unique()
user_sim_df.columns = book1.userid.unique()
```

```python
user_sim_df.iloc[0:100, 0:100]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:100, 0:100]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:100]
```
